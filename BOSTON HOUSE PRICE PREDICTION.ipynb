## HOUSE PRICE PREDICTION

```python
#IMPORTING THE DEPENDENCIES:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#IMPORTING THE BOSTON HOUSE PRICE DATASET:
house_price_dataset = sklearn.datasets.load_boston()
```

```python
print(house_price_dataset)
```

```python
#CONVERTING DATASET TO PANDAS DATAFRAME:
df = pd.DataFrame(house_price_dataset.data,columns = house_price_dataset.feature_names)
```

```python
df.head()
```

```python
#ADDING TARGET VARIABLE (Price) TO COLUMN OF DATAFRAME:
df['price']=house_price_dataset.target
```

```python
df.head()
```

```python
#CHECK SHAPE OF DATAFRAME:
df.shape
```

```python
#CHECK FOR MISSING VALUES:
df.isnull().sum()
```

```python
#STATSTICAL MEASURE OF DATASET:
df.describe()
```

```python
#PLOTTING HEATMAP AND UNDERSTANDING CORRELATION:
correlation =  df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True , square=True ,fmt='.1f' , annot=True ,annot_kws={'size':8} , cmap='Blues')
```

```python
#SPLITTING THE DATA & TARGET:
X = df.drop(['price'],axis=1)
Y = df['price']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test =train_test_split(X,Y,test_size = 0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### MODEL TRAINING

XGBOOST REGRESSOR

```python
#LOADING THE MODEL
model = XGBRegressor()
```

```python
#TRAINIG THE MODEL WITH X_TRAIN
model.fit(X_train, Y_train)
```

```python
#ACCURACY EVALUATION OF TRAINING DATA:
training_prediction_data = model.predict(X_train)
print(training_prediction_data)
```

```python
# R SQUARED ERROR & MEAN ABSOLUTE ERROR
score_1 = metrics.r2_score(Y_train,training_prediction_data)
score_2 = metrics.mean_absolute_error(Y_train,training_prediction_data)

print("R SQUARED ERROR: ",score_1)
print("Mean Absolute Error:",score_2)
```

### VISUALIZATION 

```python
plt.scatter(Y_train,training_prediction_data)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("ACTUAL VS PREDICTED - PRICE")
plt.show()
```

```python
#ACCURACY FOR PREDICTION ON TEST DATA
test_prediction_data = model.predict(X_test)
```

```python
print(test_prediction_data)
```

```python
## R SQUARED ERROR & MEAN ABSOLUTE ERROR
score_3 = metrics.r2_score(Y_test, test_prediction_data)
score_4 = metrics.mean_absolute_error(Y_test, test_prediction_data)

print("R squared Error:",score_3)
print("Mean Absolute Error:",score_4)
```

```python
plt.scatter(Y_test, test_prediction_data)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```
